## Dependencies and starter code

```python
# Dependencies and Setup
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata, study_results)
combined_data.head(50)
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_stats_table = pd.DataFrame(combined_data.groupby("Drug Regimen").count())

summary_stats_table["Mean"] = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean())
summary_stats_table["Median"] = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].median())
summary_stats_table["Standard Deviation"] = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].std())
summary_stats_table["Variance"] = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].var())
summary_stats_table["SEM"] = pd.DataFrame(combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem())

summary_stats_table = summary_stats_table[["Mouse ID", "Mean", "Median", "Variance", "Standard Deviation", "SEM"]]
summary_stats_table.head(50)
```

## Bar plots

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas
graph_datapoints = summary_stats_table[["Mouse ID"]]

graph_datapoints.plot(kind="bar", figsize=(7,5), color = "maroon", legend=True)

# Set a title for the chart
plt.title("Data Points for Each Treatment Regime")

plt.show()
plt.tight_layout()
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pyplot
x_axis = np.arange(len(graph_datapoints))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(7,5))
plt.bar(x_axis, graph_datapoints["Mouse ID"], color = "maroon", width = .3)
plt.xticks(tick_locations, graph_datapoints.index.values, rotation="vertical")

# Set a Title and labels
plt.title("Data Points for Each Treatment Regimen")
plt.xlabel("Drug Regimen")

plt.show()
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers.
top_regiments = combined_data[combined_data["Drug Regimen"].isin(["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])]
top_regiments = top_regiments.sort_values(["Timepoint"], ascending=True)
top_regiments

top_regiments_data = top_regiments[["Drug Regimen", "Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]

top_regiments_data.head(50)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
top_regimens_sort = top_regiments_data.groupby(['Drug Regimen', 'Mouse ID']).last()['Tumor Volume (mm3)']
top_regimens_sort.head()

top_regimens_sort_df = top_regimens_sort.to_frame()
top_regimens_sort_df

best_4 = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']

last_df = top_regimens_sort_df.reset_index()
list_of_tumor = last_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].apply(list)
list_of_tumor_df = pd.DataFrame(list_of_tumor)
list_of_tumor_df = list_of_tumor_df.reindex(best_4)
tumor_volumes = [vol for vol in list_of_tumor_df['Tumor Volume (mm3)']]
plt.boxplot(tumor_volumes, labels=best_4)
plt.ylim(10, 80)
plt.show()
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
time_v_tumer = combined_data[combined_data["Mouse ID"].isin(["m957"])]
time_v_tumer

time_v_tumer_data = time_v_tumer[["Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]
time_v_tumer_data

plot_line_df = time_v_tumer_data.reset_index()
plot_line_df

plot_line_final = plot_line_df[["Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]
plot_line_final

lines = plot_line_final.plot.line()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
cap_scatter = combined_data[combined_data["Drug Regimen"].isin(["Capomulin"])]

cap_scatter_df = top_regiments[["Mouse ID","Weight (g)", "Tumor Volume (mm3)"]]

cap_sorted = cap_scatter.sort_values(["Weight (g)"], ascending=True)

cap_scatter_plot = cap_scatter.reset_index()

cap_grouped_weight = cap_scatter_plot.groupby("Weight (g)")["Tumor Volume (mm3)"].mean()

cap_grouped_plot = pd.DataFrame(cap_grouped_weight).reset_index()

capomulin_scatter = cap_grouped_plot.plot(kind='scatter', x='Weight (g)', y='Tumor Volume (mm3)', grid = True, figsize= (9,9))
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
mouse_weight = cap_scatter_df.groupby(cap_scatter_df["Mouse ID"])["Weight (g)"].mean()
tumor_volume = cap_scatter_df.groupby(cap_scatter_df["Mouse ID"])["Tumor Volume (mm3)"].mean()

slope, int, r, p, std_err = st.linregress(mouse_weight, tumor_volume)
fit = slope * mouse_weight + int

plt.scatter(mouse_weight,tumor_volume)
plt.xlabel("Weight of Mouse")
plt.ylabel("Tumor Volume")
plt.plot(mouse_weight,fit,"--")
plt.xticks(mouse_weight, rotation=90)
plt.show()

# Caclc coefficient
corr = round(st.pearsonr(mouse_weight,tumor_volume)[0],2)
print(f'The correlation between weight & tumor value is: {corr}')
```
